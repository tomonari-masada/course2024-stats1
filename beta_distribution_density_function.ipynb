# ベータ分布の確率密度関数

$$p(x; a,b) = \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}x^{a-1}(1-x)^{b-1}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

def plot_beta_distribution(a, b):
  x = np.linspace(0, 1, 1000)
  y = beta.pdf(x, a, b)

  plt.figure(figsize=(10, 6))
  plt.plot(x, y, 'b-', lw=2, label=f'Beta({a}, {b})')

  plt.title(f'Beta Distribution Density Function (a={a}, b={b})')
  plt.xlabel('x')
  plt.ylabel('Density')
  plt.legend()
  plt.grid(True, alpha=0.3)

  plt.show()
```

```python
# 具体例
a = 2
b = 5
plot_beta_distribution(a, b)
```
